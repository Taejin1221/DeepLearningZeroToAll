```python
import tensorflow as tf
```

```python
x1_data = [ 73., 93., 89., 96., 73. ]
x2_data = [ 80., 88., 91., 98., 66. ]
x3_data = [ 75., 93., 90., 100., 70. ]

y_data = [ 152., 185., 180., 96., 142. ]
```

```python
w1 = tf.Variable( tf.random.normal( [ 1 ] ), name = 'weight1' )
w2 = tf.Variable( tf.random.normal( [ 1 ] ), name = 'weight2' )
w3 = tf.Variable( tf.random.normal( [ 1 ] ), name = 'weight3' )
b = tf.Variable( tf.random.normal( [ 1 ] ), name = 'bias' )
```

```python
def Hypothesis( X1, X2, X3 ):
    return  X1 * w1 + X2 * w2 + X3 * w3 + b

def Cost( X1, X2, X3, Y ):
    return tf.reduce_mean( tf.square( Hypothesis( X1, X2, X3 ) - Y ) )

def Minimize( X1, X2, X3, Y ):
    loss = lambda : Cost(X1, X2, X3, Y)
    tf.keras.optimizers.SGD( 1e-5 ).minimize( loss, [ w1, w2, w3, b ] )
```

```python
X1 = x1_data
X2 = x2_data
X3 = x3_data
Y = y_data

for step in range( 2001 ):
    cost_val, hy_val = Cost( X1, X2, X3, Y ), Hypothesis( X1, X2, X3 )
    Minimize( X1, X2, X3, Y )
    if step % 10 == 0:
        tf.print( step, 'Cost: ', cost_val, '\nPrediction:\n', hy_val)
```
