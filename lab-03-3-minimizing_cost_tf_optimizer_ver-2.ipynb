```python
import tensorflow as tf
```

```python
X = [ 1, 2, 3 ]
Y = [ 1, 2, 3 ]
```

```python
W = tf.Variable( 5.0 )
```

```python
def Hypothesis( X ):
    return X * W

def Cost( X, Y ):
    return ( tf.reduce_sum( tf.square( Hypothesis( X ) - Y ) ) )

def Minimize( X, Y ):
    loss = lambda : Cost( X, Y )
    
    tf.keras.optimizers.SGD( 0.01 ).minimize( loss, [ W ] )
```

```python
for step in range( 101 ):
    Minimize( X, Y )
    tf.print( step, W)
```
