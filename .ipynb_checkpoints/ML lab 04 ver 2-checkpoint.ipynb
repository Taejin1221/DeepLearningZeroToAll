```python
import tensorflow as tf
```

```python
x_data = [
    [ 73., 80., 75. ],
    [ 93., 88., 93. ],
    [ 89., 91., 90. ],
    [ 96., 98., 100. ],
    [ 73., 66., 70. ]
]
y_data = [
    [ 152. ],
    [ 185. ],
    [ 180. ],
    [ 196. ],
    [ 142. ]
]
```

```python
W = tf.Variable( tf.random.normal( [ 3, 1 ] ), name = 'W' )
b = tf.Variable( tf.random.normal( [ 1 ] ), name = 'b' )
```

```python
def Hypothesis(X):
    return tf.matmul( X, W ) + b

def Cost(X, Y):
    return tf.reduce_mean(tf.square( Hypothesis( X ) - Y ) )

def Minimize(X, Y):
    loss = lambda : Cost(X, Y)
    tf.keras.optimizers.SGD( 1e-5 ).minimize( loss, [ W, b ] )
```

```python
X = x_data
Y = y_data

for step in range( 2001 ):
    cost_val, hy_val = Cost( X, Y ), Hypothesis( X )
    Minimize( X, Y )
    if step % 10 == 0:
        tf.print( step, 'Cost: ', cost_val, '\nPrediction:\n', hy_val)
```
