```python
import tensorflow as tf
import numpy as np
```

```python
x_data = np.array( [
    [ 0, 0 ],
    [ 0, 1 ],
    [ 1, 0 ],
    [ 1, 1 ],
], dtype = np.float32 )

y_data = np.array( [
    [ 0 ],
    [ 1 ],
    [ 1 ],
    [ 0 ]
], dtype = np.float32 )
```

```python
W = tf.Variable( tf.random.normal( [ 2, 1 ] ), name = 'weight' )
b = tf.Variable( tf.random.normal( [ 1 ] ), name = 'bias' )
```

```python
@tf.function
def Hypothesis( X ):
    return ( tf.math.sigmoid( tf.matmul( X, W ) + b ) )
@tf.function
def Cost( X, Y ):
    hypothesis = Hypothesis( X )
    former = Y * tf.math.log( hypothesis )
    latter = tf.subtract( 1., Y ) * tf.math.log( 1. - hypothesis )
    return ( -tf.reduce_mean( former + latter ) )

def Minimize( X, Y ):
    loss = lambda: Cost( X, Y )
    
    tf.keras.optimizers.SGD( 0.1 ).minimize( loss, [ W, b ] )
    
def Predict( X ):
    return ( tf.cast( Hypothesis( X ) > 0.5, dtype = tf.float32 ) )

def Accuracy( X, Y ):
    return ( tf.reduce_mean( tf.cast( tf.equal( Predict( X ), Y ), dtype = tf.float32 ) ) )
```

```python
for step in range( 10001 ):
    Minimize( x_data, y_data )
    if ( step % 100 == 0 ):
        tf.print( step, Cost( x_data, y_data ), W )
```

```python
X, Y = x_data, y_data
h, c, a = Hypothesis( X ), Predict( X ), Accuracy( X, Y )
tf.print( h, c, a, sep = '\n' )
```
