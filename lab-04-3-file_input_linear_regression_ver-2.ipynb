```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt( 'Data/data-01-test-score.csv', delimiter = ',', dtype = np.float32 )
x_data = xy[ :, 0 : -1 ]
y_data = xy[ :, [ -1 ] ]
```

```python
print( x_data.shape, x_data, len( x_data ) )
print( y_data.shape, y_data )
```

```python
W = tf.Variable( tf.random.normal( [ 3, 1 ] ), name = 'weight' )
b = tf.Variable( tf.random.normal( [ 1 ] ), name = 'bias' )
```

```python
@tf.function
def Hypothesis( X ):
    return tf.matmul( X, W ) + b

@tf.function
def Cost( X, Y ):
    return tf.reduce_mean( tf.square( Hypothesis( X ) - Y ) )

def Minimize( X, Y ):
    loss = lambda : Cost( X, Y )
    
    optimizer = tf.keras.optimizers.SGD( learning_rate = 1e-5 )
    optimizer.minimize( loss, [ W, b ] )
```

```python
X = x_data
Y = y_data
for step in range( 2001 ):
    cost_val, hy_val, _ = \
        Cost( X, Y ), Hypothesis( X ), Minimize( X, Y )
    
    if step % 10 == 0:
        tf.print( step, 'Cost: ', cost_val,
             '\nPrediction: \n', hy_val )
```

```python
tf.print( 'Your score will be ',
         Hypothesis( [ [ 100., 70., 101. ] ] ) )
tf.print( 'Other scores will be ',
         Hypothesis( [ [ 60., 70., 110. ], [ 90., 100., 80. ] ] ) )
```
