```python
import tensorflow as tf
```

```python
x_data = [
    [ 73., 80., 75. ],
    [ 93., 88., 93. ],
    [ 89., 91., 90. ],
    [ 96., 98., 100. ],
    [ 73., 66., 70.]
]

y_data = [
    [ 152. ],
    [ 185. ],
    [ 180. ],
    [ 196. ],
    [ 142. ]
]
```

```python
W = tf.Variable( tf.random.truncated_normal( [ 3, 1 ] ) )
b = tf.Variable( 5. )
```

```python
def Hypothesis( X ):
    return ( tf.matmul( X, W ) + b )

def Diff( X, Y ):
    hypothesis = Hypothesis( X )
    return hypothesis - Y
    
def d_l1( X, Y ):
    return Diff( X, Y )

d_b = lambda X, Y: d_l1( X, Y )

d_w = lambda X, Y: tf.matmul( tf.transpose( X ) ,d_l1( X, Y )  )
```

```python
learning_rate = 1e-6
def Step( X, Y ):
    return [ W.assign( W - learning_rate * d_w( X, Y) ),
       b.assign( b - learning_rate * tf.reduce_mean( d_b( X, Y ) ) ) ]

def RMSE( X, Y ):
    return ( tf.reduce_mean( tf.square( ( Y - Hypothesis( X ) ) ) ) )
```

```python
for i in range( 1000 ):
    tf.print( i, Step( x_data, y_data ), RMSE( x_data, y_data ) )
```
