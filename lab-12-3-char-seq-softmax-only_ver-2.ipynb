```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
```

```python
sample = ' if you want you'
idx2char = list( set( sample ) )
char2idx = { c: i for i, c in enumerate( idx2char ) }

sample_idx = [ char2idx[ c ] for c in sample ]
x_data = [ sample_idx[ : -1 ] ]
y_data = [ sample_idx[ 1 : ] ]
```

```python
dic_size = len( char2idx )
num_classes = len( char2idx )
hidden_size = len( char2idx )
input_dim = len( char2idx )
batch_size = 1
sequence_length = len( sample ) - 1
learning_rate = 0.1
```

```python
x_one_hot = tf.keras.utils.to_categorical( x_data, num_classes )
x_for_softmax = tf.reshape( x_one_hot, [ -1, hidden_size ] )

y_one_hot = tf.keras.utils.to_categorical( y_data, num_classes )
y_for_softmax = tf.reshape( y_one_hot, [ -1, hidden_size ] )
```

```python
model = tf.keras.Sequential()
```

```python
model.add( layers.Dense( hidden_size, 'softmax', input_shape = ( input_dim, ) ) )
```

```python
optimizer = tf.keras.optimizers.Adam( learning_rate )
model.compile( optimizer, 'categorical_crossentropy', [ 'accuracy' ] )
model.fit( x_for_softmax, y_for_softmax, batch_size, 300 )
```

```python
predictions = model.predict( x_for_softmax )
result_str = []
for prediction in predictions:
    result_str.append( idx2char[ np.argmax( prediction, 0 ) ] )
    
print( 'Prediction : ', ''.join( result_str ) )
```
